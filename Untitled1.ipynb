```python
price_dict = df.groupby("room_type")["price"].median().to_dict()
print(price_dict)
df["price"] = df["price"].fillna(df["room_type"].map(price_dict))
```

```python
df['last_review'] = df['last_review'].ffill()
```

```python
missing_values2=df.isna().sum() 
missing_values2
```

```python
reviews_per_month_dict = df.groupby("room_type")["reviews_per_month"].median().to_dict()
print(reviews_per_month_dict)
df["reviews_per_month"] = df["reviews_per_month"].fillna(df["room_type"].map(reviews_per_month_dict))
```

```python
plt.figure(figsize=(10,5))
sns.set_style('whitegrid')
sns.barplot(x='room_type',y='price',data=df,errorbar=None,color='red',alpha=.5,edgecolor='black')
plt.title('Price Vs Type of room',fontsize=25,fontstyle='italic',color='red')
plt.xlabel('Room Type',fontsize=15,color='black',fontstyle='normal')
plt.ylabel('Price',fontsize=15,color='black',fontstyle='normal')
sns.despine()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x= df.reviews_per_month ,y=df.number_of_reviews, hue=df.number_of_reviews, palette='husl',s=100)
plt.ylabel('Number of Reviews')
plt.xlabel('Number of Reviews per Month')
plt.title('Number of Reviews per Month vs Number of Reviews')
plt.legend(title='Number of Reviews')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.barplot(x='room_type',y='availability_365',data=df,errorbar=None,color='pink',alpha=.5,edgecolor='black')
plt.title('vailability_365 Vs Type of room',fontsize=25,fontstyle='italic',color='pink')
plt.xlabel('Room Type',fontsize=15,color='black',fontstyle='normal')
plt.ylabel('vailability_365',fontsize=15,color='black',fontstyle='normal')
sns.despine()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="IPython")
plt.figure(figsize=(15, 6))
sns.barplot(x=df.neighbourhood, y=df.availability_365, palette='viridis')
plt.title('Availability of Rooms by Neighborhood',fontsize=35,fontstyle='italic',color='pink')
plt.xlabel('Neighborhood',fontsize=25,color='black',fontstyle='normal')
plt.ylabel('Availability',fontsize=25,color='black',fontstyle='normal')
plt.xticks(rotation=45)

plt.show()
```

```python
# رسم الرسم البياني الشريطي
plt.figure(figsize=(14, 9))
sns.barplot(data=df, x='neighbourhood', y='number_of_reviews', palette='muted')
plt.title('Distribution of Reviews by Neighbourhood',fontsize=35,color='black',fontstyle='normal')
plt.xlabel('Neighbourhood',fontsize=25,color='black',fontstyle='normal')
plt.ylabel('Number of Reviews',fontsize=25,color='black',fontstyle='normal')
plt.xticks(rotation=45)
plt.show()
```

```python
# رسم الرسم البياني الشريطي
sns.set_palette("husl")
plt.figure(figsize=(14, 8))
sns.lmplot(data=df, x='price', y='number_of_reviews', scatter_kws={"marker": "D"})
plt.title('Distribution of Reviews by Neighbourhood',fontsize=10,color='black',fontstyle='normal')
plt.xlabel('price',fontsize=10,color='black',fontstyle='normal')
plt.ylabel('Number of Reviews',fontsize=10,color='black',fontstyle='normal')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
